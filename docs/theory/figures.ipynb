```python
from graphviz import Digraph
```

```python
model_arch = Digraph(graph_attr={"rankdir": "LR"})
model_arch.node("x", "x", shape="circle")
model_arch.node("u", "u", shape="circle")
model_arch.node("z", "z", shape="circle")
model_arch.node("s", "s", shape="circle")
model_arch.node("r", "r", shape="circle")
model_arch.node("d", "z'", shape="circle")

model_arch.node("E", "Encode", shape="box")
model_arch.node("F", "Feature", shape="box")
model_arch.node("P", "Process", shape="box")
model_arch.node("C", "Compose", shape="box")
model_arch.node("D", "Decode", shape="box")
model_arch.node("I", "Integrate", shape="box")

model_arch.edge("x", "E")
model_arch.edge("u", "E")
model_arch.edge("E", "z")

model_arch.edge("z", "F")
model_arch.edge("u", "F")
model_arch.edge("F", "s")

model_arch.edge("s", "P")
model_arch.edge("u", "P")
model_arch.edge("P", "r")

model_arch.edge("z", "C")
model_arch.edge("r", "C")
model_arch.edge("C", "d")

model_arch.edge("d", "I")
model_arch.edge("I", "z", style='dashed')

model_arch.edge("z", "D")
model_arch.edge("D", "x", style='dashed')
```
